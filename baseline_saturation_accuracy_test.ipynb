## 라이브러리

```python
import os, math, random
import pandas as pd
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from dataclasses import dataclass
import torch
from typing import Any
from transformers import (
    Qwen2_5_VLForConditionalGeneration,
    AutoProcessor,
    BitsAndBytesConfig,
    get_linear_schedule_with_warmup,
)
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from tqdm.auto import tqdm

Image.MAX_IMAGE_PIXELS = None
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)
```

## ★ 채도 설정 — 이 PC에서 사용할 줄 하나만 주석 해제하세요 ★

```python
TRAIN_DIR = "/content/train"                  # PC-A : 원본 채도 (1.0)
# TRAIN_DIR = "/content/train_saturation_1.5" # PC-B : 채도 1.5배
# TRAIN_DIR = "/content/train_saturation_2.5" # PC-C : 채도 2.5배

# 채도 태그 자동 설정
if   "2.5" in TRAIN_DIR: SAT_TAG = "sat2.5"
elif "1.5" in TRAIN_DIR: SAT_TAG = "sat1.5"
else:                     SAT_TAG = "sat1.0"

print(f"학습 이미지 폴더 : {TRAIN_DIR}")
print(f"채도 태그        : {SAT_TAG}")
```

## 기본 설정

```python
MODEL_ID   = "Qwen/Qwen2.5-VL-3B-Instruct"
IMAGE_SIZE = 384
SEED       = 42
GRAD_ACCUM = 4

BASE_DIR = "/content"

# ── 데이터 크기 설정 ──────────────────────────────────────────
# 전체 5073개 중 사용할 샘플 수 (줄일수록 빠름)
TOTAL_SAMPLES = 200   # 전체 사용량 (학습 + 검증 합계)
TRAIN_RATIO   = 0.8   # 학습:검증 비율 (0.8 → 학습 160개, 검증 40개)
# ─────────────────────────────────────────────────────────────

SAVE_DIR = f"{BASE_DIR}/model_{SAT_TAG}"

random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)
```

## 데이터 로드 & 분할

train.csv 를 학습용과 검증용으로 나눕니다.  
검증 이미지는 채도 설정과 무관하게 **항상 원본(train/)** 을 사용합니다.

```python
train_df = pd.read_csv(f"{BASE_DIR}/train.csv")

# 전체에서 TOTAL_SAMPLES 개만 추출
sampled_df = train_df.sample(n=TOTAL_SAMPLES, random_state=SEED).reset_index(drop=True)

split        = int(TOTAL_SAMPLES * TRAIN_RATIO)
train_subset = sampled_df.iloc[:split].reset_index(drop=True)
valid_subset = sampled_df.iloc[split:].reset_index(drop=True)

# 학습 이미지 경로 → 선택한 채도 폴더로 교체
train_subset["path"] = train_subset["path"].apply(
    lambda p: os.path.join(TRAIN_DIR, os.path.basename(p))
)

# 검증 이미지 경로 → 항상 원본 train/ 폴더 사용
ORIG_TRAIN_DIR = f"{BASE_DIR}/train"
valid_subset["path"] = valid_subset["path"].apply(
    lambda p: os.path.join(ORIG_TRAIN_DIR, os.path.basename(p))
)

print(f"전체 샘플  : {TOTAL_SAMPLES}개")
print(f"학습       : {len(train_subset)}개  ({TRAIN_DIR})")
print(f"검증       : {len(valid_subset)}개  ({ORIG_TRAIN_DIR})  ← 항상 원본 채도")
```

## 프롬프트

```python
SYSTEM_INSTRUCT = (
    "You are a helpful visual question answering assistant. "
    "Answer using exactly one letter among a, b, c, or d. No explanation."
)

def build_mc_prompt(question, a, b, c, d):
    return (
        f"{question}\n"
        f"(a) {a}\n(b) {b}\n(c) {c}\n(d) {d}\n\n"
        "정답을 반드시 a, b, c, d 중 하나의 소문자 한 글자로만 출력하세요."
    )
```

## Dataset & Collator

```python
class VQAMCDataset(Dataset):
    def __init__(self, df, processor, train=True):
        self.df        = df.reset_index(drop=True)
        self.processor = processor
        self.train     = train

    def __len__(self): return len(self.df)

    def __getitem__(self, i):
        row = self.df.iloc[i]
        img = Image.open(row["path"]).convert("RGB")

        user_text = build_mc_prompt(
            str(row["question"]),
            str(row["a"]), str(row["b"]), str(row["c"]), str(row["d"])
        )
        messages = [
            {"role": "system", "content": [{"type": "text", "text": SYSTEM_INSTRUCT}]},
            {"role": "user",   "content": [
                {"type": "image", "image": img},
                {"type": "text",  "text":  user_text},
            ]},
        ]
        if self.train:
            gold = str(row["answer"]).strip().lower()
            messages.append({"role": "assistant",
                             "content": [{"type": "text", "text": gold}]})

        return {"messages": messages, "image": img}


@dataclass
class DataCollator:
    processor: Any
    train: bool = True

    def __call__(self, batch):
        texts, images = [], []
        for sample in batch:
            texts.append(self.processor.apply_chat_template(
                sample["messages"], tokenize=False, add_generation_prompt=False
            ))
            images.append(sample["image"])

        enc = self.processor(
            text=texts, images=images, padding=True, return_tensors="pt"
        )
        if self.train:
            enc["labels"] = enc["input_ids"].clone()
        return enc
```

## 모델 & Processor 로드

```python
processor = AutoProcessor.from_pretrained(
    MODEL_ID,
    min_pixels=IMAGE_SIZE * IMAGE_SIZE,
    max_pixels=IMAGE_SIZE * IMAGE_SIZE,
    trust_remote_code=True,
)

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)
base_model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
    MODEL_ID,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True,
)
base_model = prepare_model_for_kbit_training(base_model)
base_model.gradient_checkpointing_enable()

lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    lora_dropout=0.05,
    bias="none",
    target_modules=["q_proj","k_proj","v_proj","o_proj",
                    "gate_proj","up_proj","down_proj"],
    task_type="CAUSAL_LM",
)
model = get_peft_model(base_model, lora_config)
model.print_trainable_parameters()
```

## 학습

```python
train_ds     = VQAMCDataset(train_subset, processor, train=True)
train_loader = DataLoader(train_ds, batch_size=1, shuffle=True,
                          collate_fn=DataCollator(processor, True), num_workers=0)

optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)
num_steps = math.ceil(len(train_loader) / GRAD_ACCUM)
scheduler = get_linear_schedule_with_warmup(
    optimizer, int(num_steps * 0.03), num_steps
)
scaler = torch.cuda.amp.GradScaler(enabled=True)

model.train()
for epoch in range(1):
    running = 0.0
    pbar = tqdm(train_loader, desc=f"[{SAT_TAG}] Epoch {epoch+1} train", unit="batch")
    for step, batch in enumerate(pbar, start=1):
        batch = {k: v.to(device) for k, v in batch.items()}
        with torch.cuda.amp.autocast(dtype=torch.bfloat16):
            loss = model(**batch).loss / GRAD_ACCUM

        scaler.scale(loss).backward()
        running += loss.item()

        if step % GRAD_ACCUM == 0:
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad(set_to_none=True)
            scheduler.step()
            pbar.set_postfix({"loss": f"{running/GRAD_ACCUM:.3f}"})
            running = 0.0

print(f"\n✅ [{SAT_TAG}] 학습 완료")
```

## 정확도 측정

검증셋(원본 채도)으로 예측 후 정답과 비교해 정확도를 바로 출력합니다.

```python
def extract_choice(text: str) -> str:
    text  = text.strip().lower()
    lines = [l.strip() for l in text.splitlines() if l.strip()]
    if not lines: return "a"
    last = lines[-1]
    if last in ["a","b","c","d"]: return last
    for tok in last.split():
        if tok in ["a","b","c","d"]: return tok
    return "a"

model.eval()
preds, golds = [], []

for i in tqdm(range(len(valid_subset)), desc=f"[{SAT_TAG}] 정확도 측정"):
    row = valid_subset.iloc[i]
    img = Image.open(row["path"]).convert("RGB")   # 항상 원본 채도

    user_text = build_mc_prompt(
        row["question"], row["a"], row["b"], row["c"], row["d"]
    )
    messages = [
        {"role": "system", "content": [{"type": "text", "text": SYSTEM_INSTRUCT}]},
        {"role": "user",   "content": [
            {"type": "image", "image": img},
            {"type": "text",  "text":  user_text},
        ]},
    ]
    text   = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = processor(text=[text], images=[img], return_tensors="pt").to(device)

    with torch.no_grad():
        out_ids = model.generate(
            **inputs, max_new_tokens=2, do_sample=False,
            eos_token_id=processor.tokenizer.eos_token_id,
        )
    pred = extract_choice(processor.batch_decode(out_ids, skip_special_tokens=True)[0])
    preds.append(pred)
    golds.append(str(row["answer"]).strip().lower())

correct  = sum(p == g for p, g in zip(preds, golds))
total    = len(golds)
accuracy = correct / total * 100

print("\n" + "="*45)
print(f"  [{SAT_TAG}] 검증 결과")
print("="*45)
print(f"  학습 채도  : {SAT_TAG}")
print(f"  검증 샘플  : {total}개")
print(f"  정답 수    : {correct}개")
print(f"  정확도     : {accuracy:.2f}%")
print("="*45)
```

## 결과 저장

```python
# 예측 결과 저장 (다른 PC 결과와 취합해 비교할 때 사용)
result_df = valid_subset[["id", "answer"]].copy()
result_df["pred"]    = preds
result_df["correct"] = [p == g for p, g in zip(preds, golds)]

result_path = f"{BASE_DIR}/accuracy_{SAT_TAG}.csv"
result_df.to_csv(result_path, index=False)
print(f"💾 결과 저장: {result_path}")
print(f"   → 다른 PC 결과와 함께 accuracy_sat1.0.csv / sat1.5 / sat2.5 를 비교하세요")
```
